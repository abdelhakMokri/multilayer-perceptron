```python
import numpy as np
from numpy import loadtxt
from matplotlib import pyplot as plt
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn import metrics
```

# importation de donn√©es

```python
#on import les donnees on utlisant la method loadtxt de numpy ou read_csv de pandas

dataset = loadtxt('diabete1.csv', delimiter=',')

#on separe les donnees, X: represent les input et Y : les output

X = dataset[:,0:8]
y = dataset[:,8]
```

```python
# on  cree un objet Sequential
# on utulise Sequential pour specifier les couches et les mothode utiliser pour ce model

model = Sequential()

# on utilse Dense dans un PMC ou tout les neurone sont connecter
# 8 neurone d'entrer(variable reelles), 12 neurone cachee, 
model.add(Dense(12, input_dim=8, activation='linear'))
#fonction d'activation
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# calculer loss
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
#commence l'entrainement
p = model.fit(X, y, epochs=1000, batch_size=10)
```

```python
#evaluation de l'apprentissage
_, accuracy = model.evaluate(X, y)
print('Precision: %.2f' % (accuracy*100))
```

# test la base diabets2

```python
dataset = loadtxt('diabete2.csv', delimiter=',')
X = dataset[:,0:8]
y = dataset[:,8]

#on utilise la fonction predict_classes sur les input pr avoir les prediction
predictions = model.predict_classes(X)

#boucle qui affiche les inpute avec le resultat predicter et le resltat esperer
for i in range(len(X)):
	print('%s => %d (expected %d)' % (X[i].tolist(), predictions[i], y[i]))
```

```python
#decrire les performances de la classification
#on utilse la fonction confusion_natrix de scikit learn
#elle prend comme entree les prediction et les resultat esperer
performance_table = metrics.confusion_matrix(predictions, y)
performance_table
```

```python
#a partire de la matrice de perfermance on peut exctraire:

VP = performance_table[1][1]
VN = performance_table[0][0]
FP = performance_table[0][1]
FN = performance_table[1][0]

print('Vrais positifs:' , VP)
print('Vrais negatifs:' , VN)
print('Faux positifs:' , FP)
print('Faux negatifs:' , FN)
```

```python
#la precision de la classification: 

print((VP + VN) / float(VP + VN + FP + FN))

#methode predifine pour le calcule de la precision de la classification: 
print(metrics.accuracy_score(predictions, y))
```

```python
#erreur de la classification

print((FP + FN) / float(VP + VN + FP + FN) ) 

#ou 1 - calcule a partire de la fonction predifinie

print(1 - metrics.accuracy_score(predictions, y))
```

```python
#Rappel
print(VP / float(FN + VP))

#calculer rappel avc fonction predifine
print(metrics.recall_score(predictions, y))
```

```python
#precesion

print( VP / float(VP + FP))

print(metrics.precision_score(predictions, y))
```
